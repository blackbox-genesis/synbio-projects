```python
import tellurium as te
import matplotlib.pyplot as plt

model = te.loada("""
model NOT_Gate()
// Species
species A, C
// Parameters
k = 10; // max production rate of c
K = 5; // repression threshold
n = 2; // hill coefficient
d = 1; // degradation rate
// Default values
A = 0
C = 0
// C is repressed by A
J1: -> C ; k/(1+(A/K)^n)
J2: C -> ; d*C
end
""")

model.reset()
model.A=0
r1 = model.simulate(0, 50, 500)

model.reset()
model.A = 5
r2 = model.simulate(0, 50, 500)

model.reset()
model.A = 10
r3 = model.simulate(0, 50, 500)

plt.plot(r1[:,0], r1[:,2],label="A=0 (NO REPRESSION)")
plt.plot(r2[:,0], r2[:,2], label="A=5 (PARTIAL REPRESSION)")
plt.plot(r3[:,0], r3[:,2], label="A=10 (FULL REPRESSION)")
plt.xlabel("Time")
plt.ylabel("C concentration")
plt.title("NOT Gate Simulation")
plt.legend()
plt.show()
```
