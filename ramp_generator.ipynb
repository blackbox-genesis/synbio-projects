```python
import tellurium as te

model = te.loada("""
model ramp_generator
// Species
species Input = 0; GFP = 0;
// Parameters
k1 = 1 // input accumulation rate
k2 = 0.1 // GFP production rate
d = 0.01 // GFP degradation rate
                 
// Reactions
J1: -> Input; k1 // Constantly rising input
J2: -> GFP; k2*Input // Input triggers GFP production
J3: GFP -> ; d*GFP // GFP decay
                 
end
""")

result = model.simulate(0, 100, 1000)
model.plot(title="Ramp Generator Cicruit")
```
