```python
import tellurium as te
import matplotlib.pyplot as plt
model = te.loada("""
model togggle_switch()
 // Species
 species A, B
 // Parameters
 k1 = 50; k2 = 50
 dA = 1; dB = 1
 K = 10; n = 2
 // Reactions
 J1: -> A; k1/(1+(B/K)^n)
 J2: -> B; k2/(1+(A/K)^n)
 J3: A -> ; dA*A
 J4: B -> ; dB*B
 end
 """)

model.reset()
model.A = 0.01
model.B = 10 

result = model.simulate(0, 100, 1000)
model.plot(result, title="Toggle Switch Simulation")



                   
```
